```python
close_avg = df_google["close"].values
```

```python
# Create the moving average window
window_size = 300
window = np.ones(window_size) / window_size 
```

```python
# Apply the moving average window
closed = np.convolve(close_avg, window, mode="same")
```

```python
plt.figure(figsize=(16, 8)) 
plt.plot(df_google["date"], closed, label="Smoothed") 
plt.xlabel("date") 
plt.ylabel("close") 
plt.legend() 
plt.show()
```
